## Concat

```python
import numpy as np
import pandas as pd
```

```python
sozluk1 = {"isim":["ahmet","mehmet","zeynep","atil"],
           "spor":["koşu","yüzme","Koşu",np.nan],
           "kalori":[100,200,300,400]
           }
sozluk1 = pd.DataFrame(sozluk1,index=[1,2,3,4])
sozluk1
```

```python
sozluk2 = {"isim":["Osman","ali","veli","ata"],
           "spor":["koşu","yüzme","Koşu",np.nan],
           "kalori":[50,100,150,250]
           }
sozluk2 = pd.DataFrame(sozluk2,index=[1,2,3,4])
sozluk2
```

```python
sozluk3 = {"isim":["ahmet","mahmut","duygu","nur"],
           "spor":["koşu","yüzme","Koşu",np.nan],
           "kalori":[100,200,300,400]
           }
sozluk3 = pd.DataFrame(sozluk3,index=[1,2,3,4])
sozluk3
```

### Birleştirme / concat

```python
birlestirme = pd.concat([sozluk1,sozluk2,sozluk3])
birlestirme
```

### merge

```python
merge = pd.merge(sozluk1,sozluk2,on="spor")
merge
```
