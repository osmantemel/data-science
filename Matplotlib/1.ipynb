```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
##bazı yerlerde plt.show() yazmaya gerek kalmaz 
```

* genelde pyplot numpy üzerinden olur

```python
yasList=[10,20,30,40,50,60,70]
kiloList=[10,20,35,45,55,65,70]

numpyYasList = np.array(yasList)
numpyKiloList = np.array(kiloList)

print(numpyYasList)
print(numpyKiloList)
```

```python
plt.plot(numpyYasList,numpyKiloList,"g") # g green çizgi demek

plt.xlabel("age")
plt.ylabel("weight")
plt.title("age vs weight")
# plt.show()
```

```python
dizi = np.linspace(0,10,20)
dizi2= dizi ** 3
print(dizi2)
print(dizi)
```

```python
plt.plot(dizi,dizi2,"r--")
plt.plot(dizi,dizi2,"r*-")
plt.plot(dizi,dizi2,"r^-")
plt.xlabel("x")
plt.ylabel("y")
plt.title("x vs y")
```

# Subplot 
* grafik birleştirme

```python
plt.subplot(1,2,1) # 1 sıra 2 kolon olcak şuan 1. grafiği çiziyorum
plt.plot(dizi,dizi2,"r--")
plt.subplot(1,2,2)
plt.plot(dizi,dizi2,"r*-")
```

```python
(benFigure,benimEksenler)=plt.subplots(nrows=1,ncols=2)

for eksen in benimEksenler:
    eksen.plot(dizi,dizi2,"g*-")
    eksen.set_xlabel("x")
    eksen.set_ylabel("y")
    eksen.set_title("x vs y")
plt.tight_layout()
```

# Figure 
* Üst üste grafikler gösterme 

```python
figure = plt.figure()
figureaxes =figure.add_axes([0.2,0.2,0.4,0.4]) #0.4 ler with ve height
figureaxes.plot(dizi,dizi2,"r^-")
```

```python
figure2 = plt.figure()
eksen1 =figure2.add_axes([0.2,0.2,0.5,0.5])
eksen2 =figure2.add_axes([0.3,0.3,0.5,0.5])
```

```python
figure2 = plt.figure()
eksen1 =figure2.add_axes([0.2,0.2,0.9,0.9]) # 0.2 ler ise konumudur
eksen2 =figure2.add_axes([0.3,0.6,0.4,0.4])

eksen1.plot(dizi,dizi2,"g--")
eksen1.set_xlabel("x")
eksen1.set_ylabel("y")
eksen1.set_title("x vs y")

eksen2.plot(dizi2,dizi,"b*-")
eksen2.set_xlabel("x2")
eksen2.set_ylabel("y2")
eksen2.set_title("x2 vs y2")
```
