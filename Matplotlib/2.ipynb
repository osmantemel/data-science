### Ä°mport

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,90,7)
y = np.sin(x)
```

### Legend

```python
newFigure = plt.figure(figsize=(5,3))
newaxis = newFigure.add_axes([0.1,0.1,0.9,0.9])
newaxis.plot(x,y,"g^-",label="numpy dizisi sin()")
newaxis.plot(x,y*2,"r^-",label="numpy dizisi sin() * 2")
newaxis.legend(loc=0)
```

### Figure Kaydetme

```python
newFigure.savefig("figure.png")
```

### GÃ¶rsellik Ä°leri Seviye

```python
(MyFig ,Myaxis) = plt.subplots()
Myaxis.plot(x,y,color = "#3A95A8",linewidth =5)
```

```python
(MyFig ,Myaxis) = plt.subplots()
Myaxis.plot(x,y,color = "#4A95A8",linestyle=":")
Myaxis.plot(x,y-2,color = "#3155AA",linestyle="-.")
Myaxis.plot(x,y+2,color = "#3A9998",linestyle="--")
Myaxis.plot(x,y+4,color = "#3A9998", marker = "*")
```

```python

plt.style.use('_mpl-gallery')

# make data
x = np.arange(0, 10, 2)
ay = [1, 1.25, 2, 2.75, 3]
by = [1, 1, 1, 1, 1]
cy = [2, 1, 2, 1, 2]
y = np.vstack([ay, by, cy])

# plot
fig, ax = plt.subplots()

ax.stackplot(x, y)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
