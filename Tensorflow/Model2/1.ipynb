```python
max_price_indices = dataFr['price'].nlargest(139).index

dataFr.drop(max_price_indices, inplace=True)
dataFr=dataFr.drop("transmission",axis=1)
dataFr
```

```python
y = dataFr["price"].values
x = dataFr.drop("price", axis=1).values
print(y.shape)
print(x.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()


model.add(Dense(12, activation='relu' , ))
model.add(Dense(12, activation='relu' , ))
model.add(Dense(12, activation='relu' , ))
model.add(Dense(12, activation='relu' , ))

model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x=x_train, y=y_train, validation_data=(x_test,y_test),epochs=300 ,batch_size=300)
```

```python
loss = pd.DataFrame(model.history.history)
plt.figure(figsize=(5,5))
plt.plot(loss["loss"])
plt.plot(loss["val_loss"])
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
tahminDf =model.predict(x_test)
tahminDf
```

```python
mean_absolute_error(y_test, tahminDf)
```
